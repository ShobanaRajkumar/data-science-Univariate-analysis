```python
dataset["County Name"].fillna(dataset["County Name"].mode(),inplace=True)
dataset["Footnote"].fillna(dataset["Footnote"].mode(),inplace=True)
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(dataset)
```

```python
dataset[qual]
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='constant')
imp.fit(dataset[qual])
df=imp.transform(dataset[qual])
df=pd.DataFrame(df,columns=qual)
```
